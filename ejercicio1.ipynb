```python
from faker import Factory
from random import randint
import pandas as pd
import matplotlib.pyplot as plt

df = pd.DataFrame (columns=['nombre','email','ciudad'])
def personas(fake):
    for i in range(randint(25,50)):
        stuff = [fake.name(), fake.email().split('@')[1], fake.city()]
        df.loc[i] = [item for item in stuff]
             
if __name__ == "__main__":
    fake = Factory.create()
    personas(fake)
    print(df)
    df.nombre.groupby(df.email).count().plot(kind='pie',cmap='Paired')
plt.axis('equal')
```
